### utilities 파싱

```python
def extract_utilities(fee):
    try:
        utilities = fee['UtilitiesAndEssentials']
        return [u['FeeDisplayName'] for u in utilities]
    except:
        return []

df['utilities'] = df['transparentFee'].apply(extract_utilities)
```

```python
def estimate_cost(row):
    base = row['listingMinRent']
    
    # 단순 예시
    if 'High Speed Internet Access' not in row['utilities']:
        base += 25
    if 'Garbage Service' not in row['utilities']:
        base += 20
        
    return base

df['estimated_total_cost'] = df.apply(estimate_cost, axis=1)

df.head()
```

```python
df_clean = df[[
    'propertyName',
    'fullAddress',
    'listingCity',
    'listingNeighborhood',
    'listingMinRent',
    'listingMaxRent',
    'beds',
    'bathrooms',
    'location',
    'amenities',
    'transparentFee',
    'rentals'
]].copy()

df_clean.head()
```

```python
def extract_unit_types(rentals):
    try:
        return list(set([r['Beds'] for r in rentals]))
    except:
        return []

df_clean['bed_types'] = df_clean['rentals'].apply(extract_unit_types)
df_clean.head()
```

```python
df_clean['lat'] = df_clean['location'].apply(lambda x: x['latitude'])
df_clean['lng'] = df_clean['location'].apply(lambda x: x['longitude'])
```

```python
def extract_utilities(fee):
    try:
        return [u['FeeDisplayName'] for u in fee['UtilitiesAndEssentials']]
    except:
        return []

df_clean['utilities'] = df_clean['transparentFee'].apply(extract_utilities)
```

```python
def simplify_amenities(amenities):
    try:
        return [item for group in amenities for item in group['value']]
    except:
        return []

df_clean['amenities_simple'] = df_clean['amenities'].apply(simplify_amenities)
```

```python
def has_utility(row, keyword):
    return any(keyword.lower() in u.lower() for u in row['utilities'])
```

```python
def estimate_total_cost(row):
    base = row['listingMinRent']
    
    # utilities 없으면 추가
    if not has_utility(row, "internet"):
        base += 25
        
    if not has_utility(row, "garbage"):
        base += 20
        
    if not has_utility(row, "recycling"):
        base += 10
        
    # electricity / gas는 보통 안 나와서 기본 추가
    base += 50  # 평균값 (대충 잡은 것)
    
    return base
```

```python
df_clean['estimated_total_cost (per person)'] = df_clean.apply(estimate_total_cost, axis=1)
```

```python
df_clean['bath_min'] = df_clean['bathrooms'].apply(lambda x: x['min'] if isinstance(x, dict) else None)
df_clean['bath_max'] = df_clean['bathrooms'].apply(lambda x: x['max'] if isinstance(x, dict) else None)

df_clean = df_clean.drop(columns=['bathrooms'], errors='ignore')
```
